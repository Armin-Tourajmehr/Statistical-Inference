In this question, we will learn more about Monte Carlo simulation and understand how it can help us to model and analyze complex systems, where analytical or closed-form solutions are difficult. We are going to estimate the value of ğœ‹ using Monte Carlo simulation. The idea is to randomly generate points within a square and determine how many fall in a circle inscribed by the square. By comparing the points inside the circle to the total number of points you can approximate ğœ‹. Repeat your simulation with different number of points and analyze their results.


**<font color='green' size=5px>Answer ğŸ‘‡:**</font>

Before start coding, We know something about areas of square and circle:

- **Area of square**:
  - **A = $a^2$**

- **Area of circle**:
  - **A = Ï€$r^2$**

Suppose that we have a circle that has a radius that is half the length of the side of the square. Now we can write the area of the square as:

- **A = $(2r)^2$**
- **A = $(4r)^2$**

Now, if we know both areas we can estimate Ï€ using the formula:
- $\frac{A_c}{A_s}$ = $\frac{Ï€r^2}{4r^2}$ = $\frac{Ï€}{4}$


Now it's time to use **Monte Carlo** method and approximate both areas using some random process.

```r
estimate_pi <- function(seed = 42, points) {
  set.seed(seed)
  x <- runif(points, -1, 1) # random numbers between -1 and 1
  y <- runif(points, -1, 1)
  inside_circle = 0 # create vector
  for (i in 1:length(x)){
    dist <- sqrt(x[i]**2 + y[i]**2) # formula for distance from the origin
    if (dist < 1){
      inside_circle = inside_circle + 1
    }
  }
  pi_estimate <- (inside_circle/points) * 4 # To get Ï€ we just need to multiply it by four (because the ratio is Ï€/4)
  return(pi_estimate)
}
```

Using **sapply** to estimate **pi** for differente number of generated.

We used 6 numbers with multiples of ten to estimate pi.

```r
no_of_iterations <- c(10, 100, 1000, 10000, 100000, 1000000)
res <- sapply(no_of_iterations, function(n)estimate_pi(points = n))
results_df <- data.frame(t(res))
colnames(results_df) <- paste(no_of_iterations)
print(results_df)
```

<font color="purple" size="5px">**Analyse:**</font>

- The Monte Carlo method used in this code leverages random sampling to estimate the value of pi.
- As the number of generated points (n) increases, the accuracy of the estimate typically improves due to the law of large numbers.

<font color="orange" size="5px">**Simulation:**</font>
- First, we draw the generated points to show that by increasing the points, the circle and square can be seen more clearly.

> <font color="red" size="4px">**Note :**</font>
  - We did not draw for the number **1000000** because it would take a lot of time.

```r
estimate_pi_plot <- function(points) {
  x <- runif(points, -1, 1) # random numbers between -1 and 1
  y <- runif(points, -1, 1)
  colors = c() # create vector with colors
  for (i in 1:length(x)){
    dist <- sqrt(x[i]**2 + y[i]**2) # formula for distance from the origin
    if (dist < 1){
      color <- "red" # red if the point lies inside the circle
    } else {
      color <- "black" # black if it does not
    }
    colors <- c(colors, color)
  }
  plot(x,y, pch = 20, col =colors )
}
```
