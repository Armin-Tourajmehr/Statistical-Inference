**<font color='green' size=5px>Answer ðŸ‘‡:**</font>

```r
day_scores <- c(99,56,78,55.5,32,90,80,81,56,59,45,77,84.5,84,70,72,68,32,79,90)
evening_scores <- c(98,78,68,83,81,89,88,76,65,45,98,90,80,84.5,85,79,78,98,90,79,81,25.5)
```

**<font color='orange' size='6px'>b)</font>**  
- As shown below, there are **no outliers** for the **day class**, but there are **two outliers** with the values of <font color='blue'>**25.5**</font> and <font color='blue'>**45**</font> for the **evening class**.

**<font color='green' size=5px>Answer ðŸ‘‡:**</font>

To find outliers, we need to get IQR
- There is formula for **IQR**:
  - **IQR =**  <font color='blue'>$Q_3$ - $Q_1$</font>

Then we should get max upper and and lower whisker reach:
  - **Max upper whisker reach =** <font color='blue'>$Q_3$ + 1.5 * IQR</font>
  - **Max lower whisker reach =** <font color='blue'>$Q_1$ - 1.5 * IQR</font>

A potential **outlier** is defined as an observation beyond the maximum reach of the whiskers.

**<font color='purple' size='3px'>Outlier for Day class: </font>**

```r
# Interquartile Range
dIQR <- IQR(day_scores)
# First quartile
dfq = summary(day_scores)[2]
# Third quartile
dtq = summary(day_scores)[5]
# Calculate max lower whisker reach
dTmin = dfq - (1.5 * dIQR)
# Calculate max upper whisker reach
dTmax = dtq + (1.5 * dIQR)
# Calculate Outlier
print(paste0("The outlier for Day class is = ",day_scores[which(day_scores < dTmin | day_scores > dTmax)]))
```

**<font color='purple' size='3px'>Outlier for Evening class: </font>**

```r
# Interquartile Range
eIQR <- IQR(evening_scores)
# First quartile
efq = summary(evening_scores)[2]
# Third quartile
etq = summary(evening_scores)[5]
# Calculate max lower whisker reach
eTmin = efq - (1.5 * eIQR)
# Calculate max upper whisker reach
eTmax = etq + (1.5 * eIQR)
# Calculate Outlier
print(paste0("The outlier for Evening class is = ",evening_scores[which(evening_scores < eTmin | evening_scores > eTmax)]))
```
